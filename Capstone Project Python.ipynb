# Task-2: Identify the columns that have null values and perform the null value treatment. (Choose the imputation method based on the type of data in the columns of interest)

```python

# Check for null values in each column
null_values = df.isnull().sum()
```

```python

# Identify the columns that have null values
columns_with_null_values = null_values[null_values > 0].index
```

```python

# Perform the null value treatment based on the type of data in the columns of interest
for column in columns_with_null_values:
    if df[column].dtype == 'object':
        # Impute missing values with the most frequent value in the column
        df[column] = df[column].fillna(df[column].mode()[0])
    elif df[column].dtype == 'int64' or df[column].dtype == 'float64':
        # Impute missing values with the mean of the column
        df[column] = df[column].fillna(df[column].mean())

# Check for null values again after the treatment
null_values_after_treatment = df.isnull().sum()

# Print the columns that still have null values
columns_with_null_values_after_treatment = null_values_after_treatment[null_values_after_treatment > 0].index

print("Columns with null values before treatment:", columns_with_null_values)
print("Columns with null values after treatment:", columns_with_null_values_after_treatment)
```

# Task-10: Sort the data in descending order based on revenue and filter the top 500 movies. Find the measures of central tendency for the following columns using the filtered data:

1. budget

2. revenue

3. runtime

Perform outlier analysis for the above three columns using box plots.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset
data = pd.read_csv('DS1_C8_V3_ND_Sprint2_Data Analysis Using Python_Dataset.csv')

# Filter the top 500 movies based on revenue in descending order
filtered_data = data.sort_values(by='revenue', ascending=False).head(500)

# Calculate measures of central tendency for the filtered data
budget_mean = filtered_data['budget'].mean()
revenue_mean = filtered_data['revenue'].mean()
runtime_mean = filtered_data['runtime'].mean()

budget_median = filtered_data['budget'].median()
revenue_median = filtered_data['revenue'].median()
runtime_median = filtered_data['runtime'].median()

budget_mode = filtered_data['budget'].mode()[0]
revenue_mode = filtered_data['revenue'].mode()[0]
runtime_mode = filtered_data['runtime'].mode()[0]

# Print the measures of central tendency
print("Measures of Central Tendency:")
print("Budget:")
print(f"Mean: {budget_mean}")
print(f"Median: {budget_median}")
print(f"Mode: {budget_mode}")

print("\nRevenue:")
print(f"Mean: {revenue_mean}")
print(f"Median: {revenue_median}")
print(f"Mode: {revenue_mode}")

print("\nRuntime:")
print(f"Mean: {runtime_mean}")
print(f"Median: {runtime_median}")
print(f"Mode: {runtime_mode}")

# Perform outlier analysis for budget, revenue, and runtime using box plots
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.boxplot(filtered_data['budget'])
plt.title("Budget")

plt.subplot(1, 3, 2)
plt.boxplot(filtered_data['revenue'])
plt.title("Revenue")

plt.subplot(1, 3, 3)
plt.boxplot(filtered_data['runtime'])
plt.title("Runtime")

plt.tight_layout()
plt.show()
```
